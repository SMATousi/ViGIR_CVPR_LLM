```python
train_df = pd.read_excel(xls, sheet_name='train')
test_df = pd.read_excel(xls, sheet_name='test')
```

```python
train_classes = set(train_df['ture_class_name'].dropna().unique())
test_classes = set(test_df['ture_class_name'].dropna().unique())

# Combine unique class names from both sets
all_classes = train_classes.union(test_classes)

# Convert to a single string
class_names_string = ", ".join(sorted(all_classes))
```

```python
class_names = sorted(all_classes)
class_numbers = [i for i in range(1, len(class_names) + 1)]

# Convert class names into a string
class_names_string = ", ".join(class_names)
```
